```python
l = []
for i in df['User-Rating']:
    l.append(i[:3])
l
```

```python
df1['User-Rating'] = l
```

```python
df1['User-Rating'].head()
```

```python
df1['Singer/Artists'] = df1['Singer/Artists'].replace(' ',',')
df1['Singer/Artists'].head()
```

```python
df1['Album/Movie'] = df1['Album/Movie'].replace(' ','')
df1['Album/Movie'].head()
```

```python
df1['Tags'] = df1['Singer/Artists'] +  df1['Genre'] + df1['Album/Movie']
df1.head()
```

```python
df2 = df1[['Song-Name','Tags']]
df2.head()
```

```python
df2['Tags'] = df2['Tags'].apply(lambda x:x.lower())
df2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000)
```

```python
vector=cv.fit_transform(df2['Tags']).toarray()
vector
```

```python
vector[0]
```

```python
dir(cv)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
model = cosine_similarity(vector)
```

```python
sorted(list(enumerate(model[0])),reverse=True,key =lambda x:x[1])
```

```python
df2.rename(columns = {'Song-Name': 'Title'},inplace=True)	
```

```python
def recommend(music):
    index = df2[df2['Title'] == music].index[0]
    distance = model[index]
    music_list=sorted(list(enumerate(distance)),reverse=True,key =lambda x:x[1])
    for i in music_list:
        print(df2.iloc[i[0]]['Title'])
```

```python
recommend('Aankh Marey')
```

```python
recommend('Coca Cola')
```

```python
recommend('Tum Se')
```
